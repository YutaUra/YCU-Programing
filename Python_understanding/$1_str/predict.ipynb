```python
import xgboost as xgb
import pandas as pd
import re
import os
```

```python
pattern = re.compile(r'[\d]*.model')
model_paths = [d for d in os.listdir() if pattern.match(d)]
new_model_path = sorted(model_paths)[-1]
model = xgb.XGBRegressor()
model.load_model(new_model_path)
model
```

```python
test = pd.read_csv('test2.csv')
test.set_index('id', inplace=True)
dammy_test = pd.get_dummies(test)
dammy_test.head()
```

```python
y = model.predict(dammy_test)
```

```python
y
```

```python
pd.concat([pd.Series(test.index),pd.Series(y)],axis=1).to_csv('test_pred.csv', header=False, index=False)
```
